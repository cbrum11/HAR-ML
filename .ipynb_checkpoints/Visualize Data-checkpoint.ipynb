```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
filename = 'CLEAN_2019-06-20 16:42:57.005476'
df = pd.read_csv(filename+'.csv')
```

```python
df.head(1)
df.columns
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```

```python
plt.scatter(df.year_built, df.building_sqft)
plt.title('Year Built vs. Building Sqft');
```

```python
plt.scatter(df.list_price, df.year_built)
plt.title('List Price vs. Year Built');
```

```python
new_df = df.loc[:,['building_sqft', 'year_built', 'list_price', 'pool']]
a = sns.pairplot(data=new_df, hue='pool', height=5)
```

```python
a.savefig('seaborn1.png')
```
